## Q1. 임의의 input vector X를 numpy array로 정의하세요.
- 단, X의 column축 차원은 10개로 가정하고, 입력데이터 X의 row축 데이터는 100개로 가정합니다.

## Q2. 임의의 Weight Vector W를 numpy array로 정의하세요
- 단, W의 row축 차원은 input X의 row축 데이터의 개수이며, W의 Column축 차원은 1로 가정합니다.
- W의 값은 0~1 사이의 랜덤한 값을 갖습니다.

```python
W = np.random.random(size=(100,1))
```

```python
W.shape
```

## Q3. W와 X의 shape을 고려하여, bias vector B를 numpy array로 정의하세요
- 단, B의 element value는 0~1의 랜덤한 값을 가집니다.

```python
B = np.random.random(size = (1, 10))
```

## Q4. Matrix의 칼럼단위의 Min-Max Scaling 함수를 정의하고, Input Vector X에 대해 각 칼럼별로 Min-Max Scaling을 적용하세요
- Min-Max Scaling의 수식은 다음과 같습니다.

  <img src = "https://blog.kakaocdn.net/dn/bOOjFS/btrMp3CJGMI/clCAK4huytiEEiLDAVKgjK/img.png" height = "150px">

```python
X_sc = X.copy()
```

```python
X_sc[:, 0] -  X_sc[:, 0].max()
```

```python
def min_max_scaler(x_col):
  x_max = x_col.max()
  x_min = x_col.min()

  x_col_scaled = (x_col - x_min) / (x_max - x_min)
  return x_col_scaled

```

```python
X_sc = X_sc.astype("float")
```

```python
for i in range(X_sc.shape[1]):
  X_sc[:, i] = min_max_scaler(X_sc[:, i])
```

```python
X_sc
```

```python
X_sc.max(), X_sc.min()
```

## Q5. Weight vector W에 대해 Transpose연산을 수행하고, Scaling을 적용한 Input X와 dot product를 구한 후 bias vector를 더해서 output vector Y를 구하세요

```python
W.T.shape
```

```python
X.shape
```

```python
Y = np.dot(W.T, X_sc) + B
```

## Q6. Sigmoid function을 정의하고, Sigmoid function에 Y를 입력하고 최종 output vector Y_final을 구해보세요
- Sigmoid function의 수식은 다음과 같습니다.
  <img src = "https://media.licdn.com/dms/image/D4D12AQGIXdSG7IJCNw/article-cover_image-shrink_600_2000/0/1694183259537?e=2147483647&v=beta&t=OtnfeqwCtKTSVrdKZdyOzNYECyLLZuEUIxkTfTQ0dS0" height = "300px">

- exponential 함수를 어떻게 사용하는지 googling을 통해 search해봅시다!

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-1 * x))
```

```python
Y_pred = sigmoid(Y)
```
