##### Model Imputer
- Correlation 직접 계산 or KNN Impute 방식 사용

###### Correlation을 계산해서 인접 data 찾아 Missing Value Impute하기

````python
df_copy_4 = df_new.copy()
````

````python
df_copy_4.info()
````

````python
numeric_cols = df_copy_4.describe().columns
non_null_data = df_copy_4.loc[df_copy_4['전용면적'].notnull(), numeric_cols.drop('전용면적')]
null_data = df_copy_4.loc[df_copy_4['전용면적'].isnull(), numeric_cols.drop('전용면적')]
````

````python
x = null_data.iloc[0].values
````

````python
target_df = non_null_data
````

````python
from sklearn.metrics.pairwise import cosine_similarity
````

````python
cosine_similarity(df_copy_4.loc[:, numeric_cols.drop('전용면적')])
````

````python
cos_sim_arr = cosine_similarity(df_copy_4.loc[:, numeric_cols.drop('전용면적')])
````

````python
cos_sim_arr[null_data.index]
````

````python
for idx in df_copy_4.loc[df_copy_4['전용면적'].isnull()].index:
  arg_sort_list =  np.argsort(cos_sim_arr[idx]).tolist()
  if arg_sort_list[-1] == idx:
    nn_index_list = arg_sort_list[-4:-1]
  else:
    nn_index_list = arg_sort_list[-3:]


  df_copy_4.loc[idx, '전용면적'] = df_copy_4.loc[nn_index_list, '전용면적'].mean()
````

````python
df_copy_4.info()
````
